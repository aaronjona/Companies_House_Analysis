```python
import networkx as nx
import pandas as pd
import numpy as np
from tqdm import tqdm
import glob
import json
import matplotlib.pyplot as plt
import re
from random import randint
import operator
```

```python
sic_table = pd.read_csv('category_df.csv', dtype = {'section_code': str, 'section': str, 'SIC_code': str, 
                                                    'description': str})
companies_1 = pd.read_csv('SIC_list_1.csv', dtype = str)
companies_2 = pd.read_csv('SIC_list_2.csv', dtype = str)
companies = pd.concat([companies_1, companies_2])
del companies_1, companies_2
```

```python
companies
```

```python
sic_table[sic_table['section_code']=='D']
```

```python
combined = companies.join(sic_table.set_index('SIC_code'), on='sic')
combined
```

```python
combined.section_code.value_counts()
```

```python
len(combined[combined['sic'] == '99999'])
```

```python
combined[combined['has_been_liquidated'] == 'True']
```

```python
all_company_details = combined[['company_status', 'CompanyName', 'CompanyNumber', 'registered_office_address', 'section_code', 'sic', 'description']]
all_company_details
```

```python
all_company_details.to_csv('Company_SIC_from_company_status.csv', index = False)
```
