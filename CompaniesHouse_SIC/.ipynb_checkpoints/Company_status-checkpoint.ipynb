```python
import pandas as pd
import glob
from tqdm import tqdm
import json
```

```python
company_id = set([i.split('/')[-2] for i in glob.glob('data/*/*') if 'company_status' in i])
```

```python
len(company_id)
```

```python
officer_status = []
for c in tqdm(company_id):
    info = json.loads(open(f'data/{c}/company_status.json').read())
    status={}
    status['company_status'] = info.get('company_status', 'no_info')
    status['CompanyName'] = info.get('company_name', 'no_info')
    status['CompanyNumber'] = info.get('company_number', 'no_info')
    status['sic'] = info.get('sic_codes', 'no_info')
    status['has_insolvency_history']= info.get('has_insolvency_history', 'no_info')
    address = info.get('registered_office_address', {})
    status['registered_office_address'] = address.get('postal_code', 'no_info')
    status['type'] = info.get('type', 'no_info')
    status['has_been_liquidated'] = info.get('has_been_liquidated', 'no_info')
    status['date_of_creation'] = info.get('date_of_creation', 'no_info')
    
    officer_status.append(status)
    
officer_status_DF = pd.DataFrame(officer_status)

officer_status_DF
```

```python
officer_status_DF = officer_status_DF.explode('sic')
officer_status_DF
```

```python
officer_status_DF.to_csv('SIC_list.csv', index=False)
```
